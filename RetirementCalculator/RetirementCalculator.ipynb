```python
from matplotlib import pyplot

# 每年储蓄金额
annual_savings = 200000

# 年利率
interest_rate = 0.04
```

```python

# 绘图 并在显示每年的余额
years = list(range(1, 31))
balances = []
balance = 0
for year in years:
    old_balance = balance
    balance = (balance + annual_savings) * (1 + interest_rate)
    diff = balance - old_balance - 200000
    balances.append(balance)
    print(f"Year {year}: ${balance:,.2f}  (Increase: ${diff:,.2f})")
pyplot.plot(years, balances)
pyplot.xlabel("Years")
pyplot.ylabel("Balance ($)")
pyplot.title("Retirement Savings Over Time")
pyplot.grid(True)
pyplot.show()
```
