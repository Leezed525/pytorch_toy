```python
import matplotlib.pyplot as plot


# 根据数组绘制折线图并在每个点上显示数字
plot.figure(figsize=(10, 6))
arr = [1,2,4,2,5,7,2,4,9,0]
plot.plot(arr)
for i in range(len(arr)):
    plot.text(i, arr[i], arr[i], ha='center', va='bottom')
plot.show()


```
