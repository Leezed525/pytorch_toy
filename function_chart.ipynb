```python
# 输出所有常见的激活函数的图像
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return 2 * sigmoid(2 * x) - 1

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def swish(x, beta=1):
    return x * sigmoid(beta * x)

def mish(x):
    return x * tanh(np.log(1 + np.exp(x)))
```
